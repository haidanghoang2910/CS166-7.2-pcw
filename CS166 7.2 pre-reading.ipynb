```python
def f(x):
    Z = 24.44321494051954
    if abs(x) > 7:
        return 0
    elif abs(x) > 3:
        return 3 * (1 - (x / 7) ** 2) ** 0.5 / Z
    elif abs(x) > 1:
        return ((3 - abs(x)) / 2 - 3/7 * 10**0.5 * ((3 - x**2 + 2*abs(x))**0.5 - 2)) / Z
    elif abs(x) > 0.75:
        return (9 - 8 * abs(x)) / Z
    elif abs(x) > 0.5:
        return (3 * abs(x) + 0.75) / Z
    else:
        return 2.25 / Z
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
import random
```

```python
x_list = np.linspace(-7,7,100000)
def proposed_func(x):
    return sts.norm.rvs(loc=x, scale = 2)
```

```python
lst_sample = []
for x in x_list:
    fx =f(x)
    next_proposed = proposed_func(x)
    fy = f(next_proposed)
    prob = min(1, fy/fx)
    if random.random() < prob:
        lst_sample.append(next_proposed)
    else:
        lst_sample.append(x)
```

```python
plt.hist(lst_sample)
```

```python
lst_100th = []
for i in lst_sample:
    if lst_sample.index(i) % 100 == 0:
        lst_100th.append(i)
```
